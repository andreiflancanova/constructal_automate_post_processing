```python
from ansys.mapdl.core.errors import MapdlRuntimeError
from ansys.mapdl.core import launch_mapdl
import re
import os
import pandas as pd
pd.set_option('display.max_colwidth', 150)
```

```python
source_dir = 'D:/constructal_automate_analysis_files/ec_2a_phi_0.3_lancanova_2024_reprocess'

subdir_arr = []
file_name_arr = []

# Lista de subdiretórios
for subdir in os.listdir(source_dir):
    caminho_completo = os.path.join(source_dir, subdir)
    print(caminho_completo)
    if os.path.isdir(caminho_completo):  # Verifica se é um diretório
        subdir_arr.append(caminho_completo)  # Adiciona à lista

print('\n')
# Extrair o nome da última pasta do primeiro subdiretório
for subdir in subdir_arr:
    file_name = os.path.basename(subdir)
    print(file_name)
    file_name_arr.append(file_name)
```

```python
MAPDL_RUN_LOCATION='D:/constructal_automate_analysis_files'
MAPDL_START_TIMEOUT=120
```

```python
mapdl = launch_mapdl(
    port=50052,
    run_location=MAPDL_RUN_LOCATION,
    nproc=4,
    override=True,
    loglevel="INFO",
    start_timeout=MAPDL_START_TIMEOUT,
    cleanup_on_exit=True
)
```

```python
analysis_dict = {}

def extract_frequency_arr_for_each_analysis(analysis_dir, analysis_file_name, mapdl):
    print(f'Dir: {analysis_dir}. Filename: {analysis_file_name}')
    mapdl.cwd(analysis_dir)
    mapdl.filname(fname=analysis_file_name, key=0)
    mapdl.resume(fname=analysis_file_name, ext = 'db')
    analysis_dict[analysis_file_name] = mapdl.post_processing.frequency_values
```

```python
for subdir, file_name in zip(subdir_arr, file_name_arr):
    extract_frequency_arr_for_each_analysis(subdir, file_name, mapdl)
print(analysis_dict)
```

```python
mapdl.exit()
```

```python
for key,value in analysis_dict.items():
    print(f'Análise: {key}')
    print(value)
    print('\n')
```

```python
buckling_penultimate_loads_dict = {key: value[-3:-1] for key,value in analysis_dict.items()}
```

```python
for key,value in buckling_penultimate_loads_dict.items():
    print(f'Análise: {key}')
    print(value)
    print('\n')
```

```python
df = pd.DataFrame([
    {
        "Análise": key,
        "Antepenúltimo Valor": value[0],
        "Penúltimo Valor": value[1]}
    for key, value in buckling_penultimate_loads_dict.items()
])
```

```python
df['Ratio'] = df['Penúltimo Valor']/df['Antepenúltimo Valor']
```

```python
df['Diff'] = df['Penúltimo Valor'] - df['Antepenúltimo Valor']
```
