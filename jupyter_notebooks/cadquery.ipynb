```python
import cadquery as cq

# Dimensões básicas
a = 100  # comprimento total (eixo x)
b = 100  # largura total (eixo y)
t_p = 5  # espessura da placa
h_s = 20  # altura dos enrijecedores
t_s = 2  # espessura dos enrijecedores
s_ls = 20  # espaçamento entre enrijecedores no eixo x
s_ts = 20  # espaçamento entre enrijecedores no eixo y

# Criar a placa de base
base_plate = cq.Workplane("XY").box(a, b, t_p)

# Criar os enrijecedores longitudinais (eixo x)
for i in range(1, int(a / s_ls)):
    base_plate = base_plate.union(
        cq.Workplane("XY").translate((i * s_ls - a / 2, 0, t_p + h_s / 2)).box(t_s, b, h_s)
    )

# Criar os enrijecedores transversais (eixo y)
for j in range(1, int(b / s_ts)):
    base_plate = base_plate.union(
        cq.Workplane("XY").translate((0, j * s_ts - b / 2, t_p + h_s / 2)).box(a, t_s, h_s)
    )

# Mostrar o modelo
display(base_plate)
```
