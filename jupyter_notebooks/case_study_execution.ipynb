```python
import mysql.connector
from mysql.connector import Error

# Função para se conectar ao banco de dados e executar a query
def buscar_stiffened_plates():
    try:
        # Conexão com o banco de dados MariaDB
        connection = mysql.connector.connect(
            host='localhost',  # Substitua pelo seu host
            port=3306,
            database='constructal_automate_results',
            user='constructal_automate',  # Substitua pelo seu usuário do banco
            password='constructal_design_2024'  # Substitua pela sua senha do banco
        )
        
        if connection.is_connected():
            cursor = connection.cursor()
            
            # Query SQL
            query = """
                SELECT id FROM csg_stiffenedplate sp
                WHERE sp.phi = 0.3
                  AND sp.N_ts IN (2, 3, 4, 5)
                  AND sp.N_ls IN (3, 4)
                  AND sp.k != 0;
            """
            
            # Executa a query
            cursor.execute(query)
            
            # Busca todos os resultados
            resultados = cursor.fetchall()
            
            # Extrai os ids para a lista stiffened_plates_to_analyse
            stiffened_plates_to_analyse = [row[0] for row in resultados]
            
            print(f" {len(stiffened_plates_to_analyse)} IDs encontrados: {stiffened_plates_to_analyse}")
            return stiffened_plates_to_analyse
            
    except Error as e:
        print(f"Erro ao conectar ao banco de dados: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()
            print("Conexão ao banco de dados fechada.")

# Chama a função
stiffened_plates_to_analyse = buscar_stiffened_plates()
```

```python
import requests
from requests.auth import HTTPBasicAuth

# Configurações da API
STIFFENED_PLATE_ANALYSIS_BASE_URL = 'http://127.0.0.1:8000/stiffened-plate-analysis/'
ELASTIC_BUCKLING_BASE_URL = 'http://127.0.0.1:8000/elastic-buckling/'
ELASTO_PLASTIC_BUCKLING_BASE_URL = 'http://127.0.0.1:8000/elasto-plastic-buckling/'
USERNAME = 'andreiflancanova'
PASSWORD = 'teste1234'

def generate_stiffened_plate_analysis(stiffened_plate_id):
    try:
        # Fazendo a requisição POST com Basic Auth
        request_body = {
            "mesh_size": 25,
            "buckling_load_type": 2,
            "case_study": "case-study-lancanova-2024",
            "material": 1,
            "stiffened_plate": stiffened_plate_id
        }
        
        response = requests.post(STIFFENED_PLATE_ANALYSIS_BASE_URL, json=request_body, auth=HTTPBasicAuth(USERNAME, PASSWORD))

        # Tratamento de erros de resposta HTTP
        if response.status_code == 201:
            response_json = response.json()  # Converte a resposta em JSON
            id = response_json.get("id")
            print(f"STIFFENED_PLATE_ANALYSIS:")
            print(f"{response_json}")
            return id
        elif response.status_code == 400:
            print(f"STIFFENED_PLATE_ANALYSIS: Erro de validação para o payload {request_body}: {response.json()}")
            return None
        elif response.status_code == 401:
            print("STIFFENED_PLATE_ANALYSIS: Falha na autenticação. Verifique suas credenciais.")
            return None
        else:
            print(f"STIFFENED_PLATE_ANALYSIS: Erro inesperado na requisição com payload {request_body}: {response.status_code} - {response.text}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"STIFFENED_PLATE_ANALYSIS: Erro de conexão ou de requisição para a stiffened_plate = {stiffened_plate_id}: {e}")
        return None

def generate_elastic_buckling(stiffened_plate_analysis_id):
    try:
        # Fazendo a requisição POST com Basic Auth
        request_body = {
            "stiffened_plate_analysis": stiffened_plate_analysis_id,
            "n_x": 1.000,
            "csi_y": 1.000
        }

        response = requests.post(ELASTIC_BUCKLING_BASE_URL, json=request_body, auth=HTTPBasicAuth(USERNAME, PASSWORD))

        # Tratamento de erros de resposta HTTP
        if response.status_code == 201:
            response_json = response.json()
            id = response_json.get("id")
            print(f"ELASTIC_BUCKLING:")
            print(f"{response_json}")
            return id
        elif response.status_code == 400:
            print(f"ELASTIC_BUCKLING: Erro de validação para o payload {request_body}: {response.json()}")
            return None
        elif response.status_code == 401:
            print("ELASTIC_BUCKLING: Falha na autenticação. Verifique suas credenciais.")
            return None
        else:
            print(f"ELASTIC_BUCKLING: Erro inesperado na requisição com payload {request_body}: {response.status_code} - {response.text}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"ELASTIC_BUCKLING: Erro de conexão ou de requisição para a stiffened_plate_analysis = {stiffened_plate_analysis_id}: {e}")
        return None
        
def generate_elasto_plastic_buckling(stiffened_plate_analysis_id):
    try:
        # Fazendo a requisição POST com Basic Auth
        request_body = {
            "stiffened_plate_analysis": stiffened_plate_analysis_id,
        }

        response = requests.post(ELASTO_PLASTIC_BUCKLING_BASE_URL, json=request_body, auth=HTTPBasicAuth(USERNAME, PASSWORD))

        # Tratamento de erros de resposta HTTP
        if response.status_code == 201:
            response_json = response.json()
            id = response_json.get("id")
            print(f"ELASTO_PLASTIC_BUCKLING:")
            print(f"{response_json}")
            return id
        elif response.status_code == 400:
            print(f"ELASTO_PLASTIC_BUCKLING: Erro de validação para o payload {request_body}: {response.json()}")
            return None
        elif response.status_code == 401:
            print("ELASTO_PLASTIC_BUCKLING: Falha na autenticação. Verifique suas credenciais.")
            return None
        else:
            print(f"Erro inesperado na requisição com payload {request_body}: {response.status_code} - {response.text}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"ELASTO_PLASTIC_BUCKLING: Erro de conexão ou de requisição para a stiffened_plate_analysis = {stiffened_plate_analysis_id}: {e}")
        return None
    print('\n')

# id_teste = generate_elastic_buckling(2)
id_teste = generate_elasto_plastic_buckling(2)
```
